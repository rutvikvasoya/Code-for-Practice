```python
# importing the all require libraries
import matplotlib.pyplot as plt
import cv2
import face_recognition
```

```python
#loading the image for detect

image_to_detect = cv2.imread("index.jpeg")       #loading the image from local


plt.imshow(image_to_detect)
```

```python
all_face_location = face_recognition.face_locations(image_to_detect,model = 'hog')
#all_face_location = face_recognition.face_locations(image_to_detect,model = 'cnn')

#model can be 'cnn' or 'hog'
#when we have to use cpu and do not have more computational power then we should use hog otherwise we use cnn


print(len(all_face_location))  #this is the number of faces in image
```

```python
all_face_location
```

```python
#now detect and capture two faces  from image

#frist we have to find location of that faces in image after this we have detect a particular portion in image




#loop throught face location

for index,current_face_location in enumerate(all_face_location):
    top_pos,right_pos,bottom_pos,left_pos = current_face_location
    print(f'found faces at top {top_pos},bottom {bottom_pos},left {left_pos}, right {right_pos}')
    
    #now detecting the particular portion in image
    current_face_image = image_to_detect[top_pos:bottom_pos,left_pos:right_pos]
    plt.imshow(current_face_image)
    plt.show()
```
