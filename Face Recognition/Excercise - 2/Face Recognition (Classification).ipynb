```python
import cv2
import face_recognition
import matplotlib.pyplot as plt
```

```python
#load the sample image and get the 128 face embedding from them

tony_image = face_recognition.load_image_file('tony.jpg')
tony_face_encoding = face_recognition.face_encodings(tony_image)[0]

cap_image = face_recognition.load_image_file('cap.jpeg')
cap_face_encoding = face_recognition.face_encodings(cap_image)[0]  #we have only one face in this image,so we give 0

know_face_encoding = [tony_face_encoding,cap_face_encoding]
know_face_names = ['tony','captain']

#load the unknow image to identify the faces
image_to_recognition = face_recognition.load_image_file('tony-cap.jpeg')

#find all faces and face encoding in the unknown faces
all_face_location  = face_recognition.face_locations(image_to_recognition,model='hog')
all_face_encoding = face_recognition.face_encodings(image_to_recognition,all_face_location)

#looping through the face location and the face embeddings

for current_face_location, current_face_encoding in zip(all_face_location,all_face_encoding):
    
    #spliting the tuple to get the four position value from the current face
    top_pos,right_pos,bottom_pos,left_pos = current_face_location
    
    
    #find the all matches and get the list of matches
    all_mathces = face_recognition.compare_faces(all_face_encoding,current_face_encoding)
    print(all_mathces)
    
    #string to hold to label
    name_of_person = "unknow"
    
    #check if the all matches have at least one item , if yes get the index number of face that is located in the first
    #index  of all matches get the name of corresponding to the index number and save it in the name_of_person
    
    if True in all_mathces:
        first_match_index = all_mathces.index(True)
        name_of_person = know_face_names[first_match_index]
    
    #now draw the rectange around around all the detected faces
    cv2.rectangle(image_to_recognition,(left_pos,top_pos),(right_pos,bottom_pos),(255,0,0),2)
                                                                                         #(r,g,b) (width of line)
    #display the name as text in the input
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(image_to_recognition,name_of_person,(left_pos,bottom_pos),font,0.5,(255,255,255),1)
    
  #  cv2.imshow("face indentify",image_to_recognition)
    

    plt.imshow(image_to_recognition)
```
