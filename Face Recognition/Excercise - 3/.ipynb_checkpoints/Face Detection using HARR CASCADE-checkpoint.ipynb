```python
import cv2
```

```python
image_to_detect = cv2.imread('index.jpeg')
```

```python
#load the pretrained haar classifier model
model = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#detect the all face location  using haar classifier
all_face_location  = model.detectMultiScale(image_to_detect)
```

```python
#print number of faces we detect from the image
print(len(all_face_location))
```

```python
#looping through the face location

for index, curren_face_location in enumerate(all_face_location):
    
    #spliting the touple of get the four position value of current face
    x, y, width , hight = curren_face_location
   
    #start co-ordinate
    left_x , left_y = x ,y
    right_x = x + width 
    right_y = y + hight

    
    
    #slicing the current face from the main image
    current_face_image = image_to_detect[left_y:right_y,left_x:right_x]
    
    #showing the current face with dynamic title
    cv2.imshow("Face No"+str(index+1),current_face_image)
    
    #draw bounding box around the faces
    cv2.rectangle(image_to_detect,(left_x,left_y),(right_x,right_y),(0,255,0),2)
    

    
cv2.waitKey(0)
#close all window
cv2.destroyAllWindows()
```
