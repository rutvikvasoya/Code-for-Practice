```python
import cv2
import dlib
```

```python
image_to_detect = cv2.imread('index.jpeg')


#create a grey scale image (optinal)
image_to_detect = cv2.cvtColor(image_to_detect, cv2.COLOR_BGR2HSV)

#load the pretrained model HOG SVN model
model = dlib.get_frontal_face_detector()

#detect all the face location using model
all_face_location  = model(image_to_detect,1)


for index, current_face_location in enumerate(all_face_location):
    #start and end co-ordinates
    left,top,right,bottom = current_face_location.left(),current_face_location.top(),current_face_location.right(),current_face_location.bottom()
    
    
    #slicing the current face from the main image
    current_face_image = image_to_detect[top:bottom,left:right]
    
    #showing the current face with dynamic title
    cv2.imshow("Face No"+str(index+1),current_face_image)
    
    #draw bounding box around the faces
    cv2.rectangle(image_to_detect,(left,top),(right,bottom),(0,255,0),2)

    
cv2.waitKey(0)
#close all window
cv2.destroyAllWindows()
```
