```python
import  matplotlib.pyplot as plt
import numpy as np
import matplotlib.style as st
```

```python
subject = ["AI","ML","DS","JAVA","PHP"]
students = [20,18,34,55,30]
studnets2 = [18,45,23,34,43]
studnets3 = [28,55,38,24,33]

```

```python
plt.bar(subject,students)
```

```python
plt.bar(subject,students , width=0.6 , color="k", edgecolor="y", linewidth=2 , linestyle= "--" , label="number of student")
st.use("ggplot")
plt.figure(figsize=(16,9))
```

```python
plt.bar(subject,students , width=0.6 , color="k", edgecolor="y", linewidth=2 , linestyle= "--" , label="number of student")
plt.bar(subject,studnets2 , width=0.6 , color="g", edgecolor="y", linewidth=2 , linestyle= "--" , label="number of student2")


st.use("ggplot")
plt.xlabel("subjects")
plt.ylabel("number of students")
```

```python
width = 0.2
classes_index = np.arange(len(subject))

plt.bar(classes_index,students , width , color="k", edgecolor="y", linewidth=2 , linestyle= "--" , label="S1")
plt.bar(classes_index+width,studnets2 , width , color="g", edgecolor="y", linewidth=2 , linestyle= "--" , label="S2")
plt.bar(classes_index+width+width,studnets3 , width , color="r", edgecolor="y", linewidth=2 , linestyle= "--" , label="S3")

plt.xticks(classes_index, subject)
st.use("ggplot")
plt.xlabel("subjects")
plt.ylabel("number of students")
plt.legend(loc=2)
```
