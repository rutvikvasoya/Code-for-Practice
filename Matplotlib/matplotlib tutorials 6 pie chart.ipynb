```python
import matplotlib.pyplot as plt
```

```python
subject = ["AI","ML","DS","Java","PhP"]
marks =  [44,55,34,67,56]

```

```python
plt.pie(marks , labels=subject)
plt.show()
```

```python
explode = [0.01,0.1,0.01,0.01,0.01]     #expload means cutout from circle
colors = ["r","g","c","b","y"]



plt.pie(marks , labels=subject , explode=explode, colors=colors , autopct="%0.2f%%" , radius=1.6 , textprops={"fontsize":15}) 
plt.show()
```
